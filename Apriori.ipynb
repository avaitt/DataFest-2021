```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

#https://michael.hahsler.net/research/recommender/associationrules.html
#https://www.kdnuggets.com/2016/04/association-rules-apriori-algorithm-tutorial.html

df = pd.read_csv('ca.csv')

df.head()
```

```python
def rules(df2):
    
    drugs = ["FENT_USE", "BUP_USE", "METH_USE", "MORPH_USE", "OXY_USE", "OXYM_USE", "TRAM_USE", "TAP_USE", "COD_USE", "HYD_USE", "HYDM_USE", "SUF_USE", "STIM_USE", "BENZ_USE", "THC_USE"]

    df3 = df2[drugs]

    frq_items = apriori(df3, min_support = 0.05, use_colnames = True)

    rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
    rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])

    return rules.sort_values(by = 'lift', ascending = False).reset_index().drop(['index'], axis = 1)
```

```python
rules(df)
```

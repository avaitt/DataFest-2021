```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

#https://michael.hahsler.net/research/recommender/associationrules.html
#https://www.kdnuggets.com/2016/04/association-rules-apriori-algorithm-tutorial.html
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
def rules_ca():
    
    df_c = pd.read_csv('ca.csv')
    
    col = []
    for i in df_c.columns:
        if i.find('_USE') > -1 and i.count('_') == 1 and len(df_c[i].unique()) == 2:
            col.append(i)
    
    col = col[:15]
    
    df_c2 = df_c[col]
    
    frq_items = apriori(df_c2, min_support = 0.05, use_colnames = True)

    rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
    rules = rules.sort_values(by = 'lift', ascending = False)
    rules['country'] = 'CA'

    return rules
```

```python
ca = rules_ca().head()
ca
```

```python
def rules_us():
    
    df_c = pd.read_csv('us.csv')
    
    col = []
    for i in df_c.columns:
        if i.find('_USE') > -1 and i.count('_') == 1 and len(df_c[i].unique()) == 2:
                col.append(i)
    
    col = col[:43]
    
    df_c2 = df_c[col]
    
    frq_items = apriori(df_c2, min_support = 0.05, use_colnames = True)

    rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
    rules = rules.sort_values(by  = 'lift', ascending = False)
    rules['country'] = 'US'

    return rules
```
